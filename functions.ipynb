```python
jobs = pd.read_csv("jobs.csv")
education = pd.read_csv("education.csv")
industries = pd.read_csv("industries.csv")
prices = pd.read_csv("prices.csv")
```

```python
def get_jobs_city_plus_industry(city, industry):
    jobs_city = jobs[jobs.city == city]
    jobs_city_industry = jobs_city[jobs_city.category_name == industry]
    return jobs_city_industry

def get_industry_educational_data(title):
    industries_title = industries[industries.NAICS_title == title]
    columns_to_delete= ['10_percent', '25_percent', '50_percent', '75_percent', '90_percent']
    for i in columns_to_delete:
        del industries_title[i]
    return industries_title
    
jobs_chicago_finance = get_jobs_city_plus_industry('Chicago', "Accounting & Finance")
industries_finance = get_industry_educational_data('Nondepository Credit Intermediation')
```
